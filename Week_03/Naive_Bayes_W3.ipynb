Naive Bayes / Gaussian Naive Bayes

## Objetivo
Resolver un ejercicio completo de clasificación con Gaussian Naive Bayes usando Iris y un ejemplo adicional con frutas.


## Bloque 1. Importar librerías

En este bloque se cargan las librerías necesarias para manipular datos, visualizarlos, entrenar el modelo y construir un ejemplo extra con frutas.


```python
# Importamos NumPy para manejar arreglos.
import numpy as np

# Importamos Matplotlib para crear gráficas.
import matplotlib.pyplot as plt

# Importamos Pandas para construir el dataset de frutas.
import pandas as pd

# Importamos el dataset Iris.
from sklearn.datasets import load_iris

# Importamos Gaussian Naive Bayes.
from sklearn.naive_bayes import GaussianNB

# Importamos la función para dividir datos.
from sklearn.model_selection import train_test_split

# Importamos métricas de evaluación.
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

### Interpretación del bloque 1

Este bloque deja listo el entorno de trabajo.

Aquí quedan preparadas las herramientas para:
- cargar datos,
- construir visualizaciones,
- entrenar el clasificador probabilístico,
- y evaluar su desempeño.


## Bloque 2. Cargar Iris y explorar su estructura

En este bloque se carga el dataset, se separan las variables y se revisa la estructura general del problema.


```python
# Cargamos el dataset Iris.
iris = load_iris()

# Guardamos las variables de entrada.
X = iris.data

# Guardamos las clases.
y = iris.target

# Mostramos dimensiones.
print("Shape de X:", X.shape)
print("Shape de y:", y.shape)

# Mostramos nombres de clases.
print("Clases:", iris.target_names)

# Mostramos nombres de variables.
print("Variables:", iris.feature_names)

# Mostramos primeras observaciones.
print("\nPrimeras 5 observaciones de X:")
print(X[:5])

# Mostramos primeras etiquetas.
print("\nPrimeras 5 etiquetas de y:")
print(y[:5])
```

### Interpretación del bloque 2

La salida indica que:
- hay 150 observaciones,
- cada una tiene 4 variables,
- y existen 3 clases.

Esto confirma que también aquí estamos ante un problema de clasificación supervisada.


## Bloque 3. Visualizar dos variables de Iris

En este bloque se usan dos variables para representar el dataset en un plano y entender mejor su distribución.


```python
# Seleccionamos dos variables para visualizar en 2D.
X_vis = X[:, 0:2]

# Creamos figura.
_, ax = plt.subplots(figsize=(7, 5))

# Dibujamos la dispersión de los datos.
ax.scatter(
    X_vis[:, 0],        # Eje X
    X_vis[:, 1],        # Eje Y
    c=y,                # Color por clase
    cmap='Set1',        # Paleta
    edgecolors='k'      # Borde negro
)

# Título de la gráfica.
ax.set_title("Distribución del dataset Iris (2 variables)")

# Etiquetas de ejes.
ax.set_xlabel(iris.feature_names[0])
ax.set_ylabel(iris.feature_names[1])

# Mostramos la figura.
plt.show()
```

### Interpretación del bloque 3

Esta gráfica permite observar la separación entre clases.

Aunque Naive Bayes no clasifica por vecinos, esta visualización sigue siendo útil porque ayuda a anticipar:
- qué clases parecen separadas,
- cuáles podrían confundirse,
- y qué estructura general tienen los datos.


## Bloque 4. Dividir datos en entrenamiento y prueba

En este bloque se separa el dataset para entrenar el modelo en una parte y evaluarlo en otra.


```python
# Dividimos el dataset en entrenamiento y prueba.
X_train, X_test, y_train, y_test = train_test_split(
    X,                  # Variables de entrada
    y,                  # Clases
    test_size=0.3,      # 30% para prueba
    random_state=42,    # Reproducibilidad
    stratify=y          # Mantener proporción de clases
)

# Mostramos dimensiones.
print("Shape de X_train:", X_train.shape)
print("Shape de X_test:", X_test.shape)
print("Shape de y_train:", y_train.shape)
print("Shape de y_test:", y_test.shape)
```

### Interpretación del bloque 4

Aquí queda definido qué datos usará el modelo para aprender y cuáles se usarán para evaluarlo.

Esto permite medir la capacidad de generalización del algoritmo.


## Bloque 5. Crear y entrenar Gaussian Naive Bayes

En este bloque se construye el modelo probabilístico y se entrena con los datos de entrenamiento.


```python
# Creamos el modelo Gaussian Naive Bayes.
gnb = GaussianNB()

# Entrenamos el modelo con los datos de entrenamiento.
gnb.fit(X_train, y_train)

# Confirmamos el entrenamiento.
print("Modelo GaussianNB entrenado correctamente")
```

### Interpretación del bloque 5

En este punto el modelo ya estimó, para cada clase:
- probabilidad previa,
- medias de las variables,
- y varianzas.

Con esa información puede calcular qué clase es más probable para una nueva observación.


## Bloque 6. Hacer predicciones y evaluar el modelo

En este bloque se predicen las clases del conjunto de prueba y se revisan métricas de desempeño.


```python
# Predecimos las clases del conjunto de prueba.
y_pred = gnb.predict(X_test)

# Calculamos accuracy.
acc = accuracy_score(y_test, y_pred)

# Calculamos la matriz de confusión.
cm = confusion_matrix(y_test, y_pred)

# Generamos el reporte de clasificación.
report = classification_report(y_test, y_pred, target_names=iris.target_names)

# Mostramos accuracy.
print("Accuracy:", acc)

# Mostramos la matriz de confusión.
print("\nMatriz de confusión:")
print(cm)

# Mostramos el reporte.
print("\nReporte de clasificación:")
print(report)
```

### Interpretación del bloque 6

Aquí se ve el rendimiento del modelo en datos no vistos.

- **Accuracy** resume el porcentaje de aciertos.
- **Matriz de confusión** muestra en qué clases se equivoca.
- **Reporte de clasificación** permite mirar precisión, recall y F1-score.

Naive Bayes suele funcionar bien incluso con supuestos simplificadores.


## Bloque 7. Visualizar fronteras de decisión en 2D

En este bloque se entrena un GaussianNB con dos variables para visualizar cómo divide el plano.


```python
# Creamos un modelo GaussianNB para las variables visuales.
gnb_vis = GaussianNB()

# Entrenamos con las dos variables seleccionadas.
gnb_vis.fit(X_vis, y)

# Definimos un margen visual.
offset = 0.2

# Calculamos rangos del plano.
x_min, x_max = X_vis[:, 0].min() - offset, X_vis[:, 0].max() + offset
y_min, y_max = X_vis[:, 1].min() - offset, X_vis[:, 1].max() + offset

# Construimos la malla.
x_grid, y_grid = np.meshgrid(
    np.linspace(x_min, x_max, 200),
    np.linspace(y_min, y_max, 200)
)

# Predecimos la clase en cada punto de la malla.
pred_grid = gnb_vis.predict(np.c_[x_grid.ravel(), y_grid.ravel()])

# Reorganizamos las predicciones.
pred_grid = pred_grid.reshape(x_grid.shape)

# Creamos figura.
_, ax = plt.subplots(figsize=(7, 5))

# Pintamos el fondo según la clase predicha.
ax.pcolormesh(
    x_grid,
    y_grid,
    pred_grid,
    cmap="Set1",
    shading="auto"
)

# Superponemos los puntos reales.
ax.scatter(
    X_vis[:, 0],
    X_vis[:, 1],
    c=y,
    cmap="Set1",
    edgecolors="k"
)

# Agregamos título y etiquetas.
ax.set_title("Fronteras de decisión con Gaussian Naive Bayes")
ax.set_xlabel(iris.feature_names[0])
ax.set_ylabel(iris.feature_names[1])

# Mostramos la figura.
plt.tight_layout()
plt.show()
```

### Interpretación del bloque 7

La gráfica muestra cómo Naive Bayes divide el espacio entre clases.

A diferencia de KNN, las fronteras tienden a ser más suaves porque la clasificación depende de distribuciones y probabilidades, no de vecinos locales.


## Bloque 8. Construir un segundo ejemplo con frutas

En este bloque se crea un dataset pequeño de frutas para mostrar que el algoritmo se puede aplicar a otro problema.


```python
# Creamos un DataFrame con frutas, peso y color.
fruit_data = pd.DataFrame({
    "fruit_name": [
        "apple", "apple", "apple", "apple",
        "orange", "orange", "orange", "orange",
        "mandarin", "mandarin", "mandarin", "mandarin"
    ],
    "weight": [150, 170, 160, 180, 140, 155, 165, 175, 80, 85, 90, 95],
    "color_score": [0.72, 0.76, 0.74, 0.78, 0.88, 0.90, 0.86, 0.89, 0.60, 0.58, 0.62, 0.61]
})

# Creamos un diccionario de etiquetas.
fruit_map = {"apple": 0, "orange": 1, "mandarin": 2}

# Convertimos el nombre de fruta a etiqueta numérica.
fruit_data["fruit_label"] = fruit_data["fruit_name"].map(fruit_map)

# Mostramos la tabla.
fruit_data
```

### Interpretación del bloque 8

Este bloque traslada el algoritmo a un problema más cotidiano.

Ahora el modelo trabajará con:
- peso,
- color,
- y tipo de fruta.

Esto ayuda a ver que la lógica de clasificación no depende solo de Iris.


## Bloque 9. Entrenar Naive Bayes con frutas

En este bloque se preparan las variables y se entrena un nuevo modelo sobre el dataset de frutas.


```python
# Seleccionamos variables de entrada del dataset de frutas.
fruits_X = np.asarray(fruit_data[["weight", "color_score"]])

# Seleccionamos clases del dataset de frutas.
fruits_y = np.asarray(fruit_data["fruit_label"])

# Creamos el modelo.
gnb_fruits = GaussianNB()

# Entrenamos con las frutas.
gnb_fruits.fit(fruits_X, fruits_y)

# Confirmamos el entrenamiento.
print("Modelo de frutas entrenado correctamente")

# Mostramos dimensiones.
print("Shape de fruits_X:", fruits_X.shape)
print("Shape de fruits_y:", fruits_y.shape)
```

### Interpretación del bloque 9

En este punto el modelo ya puede clasificar frutas nuevas a partir de sus características.

Esto confirma que el mismo esquema puede reutilizarse en distintos dominios.


## Bloque 10. Visualizar frutas y fronteras de decisión

En este bloque se visualizan las frutas y luego se dibujan las regiones que el modelo asigna a cada clase.


```python
# Creamos figura para visualizar las frutas.
_, ax = plt.subplots(figsize=(7, 5))

# Dibujamos las frutas.
ax.scatter(
    fruits_X[:, 0],      # Peso
    fruits_X[:, 1],      # Color
    c=fruits_y,          # Clase
    cmap="Set2",         # Paleta
    edgecolors="k",      # Borde negro
    s=120                # Tamaño del punto
)

# Etiquetamos cada punto con su nombre.
for i, name in enumerate(fruit_data["fruit_name"]):
    ax.annotate(name, (fruits_X[i, 0] + 1, fruits_X[i, 1] + 0.003), fontsize=8)

# Título y ejes.
ax.set_title("Distribución del dataset de frutas")
ax.set_xlabel("Weight")
ax.set_ylabel("Color score")

# Mostramos la gráfica.
plt.show()

# Definimos márgenes.
offset_x = 5
offset_y = 0.05

# Calculamos rangos.
x_min_f, x_max_f = fruits_X[:, 0].min() - offset_x, fruits_X[:, 0].max() + offset_x
y_min_f, y_max_f = fruits_X[:, 1].min() - offset_y, fruits_X[:, 1].max() + offset_y

# Construimos una malla.
x_grid_f, y_grid_f = np.meshgrid(
    np.linspace(x_min_f, x_max_f, 200),
    np.linspace(y_min_f, y_max_f, 200)
)

# Predecimos la clase de cada punto de la malla.
pred_grid_f = gnb_fruits.predict(np.c_[x_grid_f.ravel(), y_grid_f.ravel()])

# Reorganizamos la salida.
pred_grid_f = pred_grid_f.reshape(x_grid_f.shape)

# Creamos nueva figura.
_, ax = plt.subplots(figsize=(7, 5))

# Pintamos el fondo con la clase predicha.
ax.pcolormesh(
    x_grid_f,
    y_grid_f,
    pred_grid_f,
    cmap="Set2",
    shading="auto"
)

# Superponemos las frutas reales.
ax.scatter(
    fruits_X[:, 0],
    fruits_X[:, 1],
    c=fruits_y,
    cmap="Set2",
    edgecolors="k",
    s=120
)

# Título y etiquetas.
ax.set_title("Fronteras de decisión en el dataset de frutas")
ax.set_xlabel("Weight")
ax.set_ylabel("Color score")

# Mostramos la gráfica.
plt.tight_layout()
plt.show()
```

### Interpretación del bloque 10

La primera gráfica muestra dónde se ubican las frutas según peso y color.

La segunda gráfica muestra cómo el modelo extiende esa información a todo el plano:
- cada región representa la fruta más probable,
- y las fronteras separan las zonas asignadas a cada clase.


## Bloque 11. Hacer una predicción nueva con frutas

En este bloque se crea una fruta ficticia y se calcula su clase más probable junto con las probabilidades por categoría.


```python
# Creamos una nueva fruta ficticia.
nueva_fruta = np.array([[168, 0.87]])

# Predecimos la clase.
pred_fruit = gnb_fruits.predict(nueva_fruta)[0]

# Calculamos probabilidades.
prob_fruit = gnb_fruits.predict_proba(nueva_fruta)[0]

# Creamos diccionario inverso para recuperar nombres.
inverse_fruit_map = {v: k for k, v in fruit_map.items()}

# Mostramos la nueva fruta.
print("Nueva fruta:", nueva_fruta)

# Mostramos la clase predicha.
print("Clase predicha:", pred_fruit)

# Mostramos el nombre correspondiente.
print("Nombre predicho:", inverse_fruit_map[pred_fruit])

# Mostramos las probabilidades por clase.
print("Probabilidades por clase:")
for idx, prob in enumerate(prob_fruit):
    print(f"  {inverse_fruit_map[idx]}: {prob:.4f}")
```

### Interpretación del bloque 11

Aquí se observa una decisión concreta del modelo.

La clase final corresponde a la categoría con mayor probabilidad posterior.  
Las probabilidades también permiten ver si la decisión fue muy clara o si hubo incertidumbre entre varias frutas.
