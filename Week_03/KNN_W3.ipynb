# K-Nearest Neighbors (KNN)


## Bloque 1. Importar librerías

En este bloque se cargan las librerías necesarias para manejar datos, visualizarlos, entrenar el modelo y evaluarlo.


```python
# Importamos NumPy para manejar arreglos numéricos.
import numpy as np

# Importamos Matplotlib para crear gráficas.
import matplotlib.pyplot as plt

# Importamos el dataset Iris.
from sklearn.datasets import load_iris

# Importamos el clasificador KNN.
from sklearn.neighbors import KNeighborsClassifier

# Importamos la función para separar entrenamiento y prueba.
from sklearn.model_selection import train_test_split

# Importamos métricas para evaluar el modelo.
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

### Interpretación del bloque 1

Este bloque prepara el entorno de trabajo.  
Aún no hay resultados del modelo, pero ya están listas las herramientas para:
- cargar el dataset,
- visualizar los datos,
- entrenar el algoritmo,
- y medir su desempeño.


## Bloque 2. Cargar el dataset y explorar su estructura

En este bloque se carga Iris, se separan variables y clases, y se revisa cómo está organizado el problema.


```python
# Cargamos el dataset Iris.
iris = load_iris()

# Guardamos las variables de entrada en X.
X = iris.data

# Guardamos las clases en y.
y = iris.target

# Mostramos la forma de X.
print("Shape de X:", X.shape)

# Mostramos la forma de y.
print("Shape de y:", y.shape)

# Mostramos los nombres de las clases.
print("Clases:", iris.target_names)

# Mostramos los nombres de las variables.
print("Variables:", iris.feature_names)

# Mostramos las primeras cinco observaciones.
print("\nPrimeras 5 observaciones de X:")
print(X[:5])

# Mostramos las primeras cinco etiquetas.
print("\nPrimeras 5 etiquetas de y:")
print(y[:5])
```

### Interpretación del bloque 2

La salida indica que:
- hay **150 observaciones**,
- cada observación tiene **4 variables**,
- y existen **3 clases** objetivo.

Esto confirma que estamos ante un problema de **clasificación supervisada**.


## Bloque 3. Visualizar dos variables del dataset

En este bloque se seleccionan dos variables para representar visualmente el problema en un plano 2D.


```python
# Seleccionamos solo dos variables para visualizar en 2D.
X_vis = X[:, 0:2]

# Creamos figura y ejes.
_, ax = plt.subplots(figsize=(7, 5))

# Dibujamos la dispersión de los datos.
ax.scatter(
    X_vis[:, 0],        # Primera variable en el eje X
    X_vis[:, 1],        # Segunda variable en el eje Y
    c=y,                # Color según la clase
    cmap='Paired',      # Paleta de colores
    edgecolors='k'      # Borde negro
)

# Agregamos título.
ax.set_title("Distribución del dataset Iris (2 variables)")

# Asignamos nombre al eje X.
ax.set_xlabel(iris.feature_names[0])

# Asignamos nombre al eje Y.
ax.set_ylabel(iris.feature_names[1])

# Mostramos la gráfica.
plt.show()
```

### Interpretación del bloque 3

La gráfica permite observar si las clases están separadas o si se superponen.

Esto da una intuición temprana sobre:
- qué tan fácil será clasificar,
- qué clases podrían confundirse,
- y cómo se distribuyen las observaciones en el espacio.


## Bloque 4. Dividir los datos en entrenamiento y prueba

En este bloque se separan los datos para entrenar el modelo con una parte y evaluarlo con otra.


```python
# Dividimos el dataset en entrenamiento y prueba.
X_train, X_test, y_train, y_test = train_test_split(
    X,                  # Variables de entrada
    y,                  # Clases
    test_size=0.3,      # 30% para prueba
    random_state=42,    # Semilla para reproducibilidad
    stratify=y          # Mantiene proporciones de clase
)

# Mostramos dimensiones del conjunto de entrenamiento.
print("Shape de X_train:", X_train.shape)

# Mostramos dimensiones del conjunto de prueba.
print("Shape de X_test:", X_test.shape)

# Mostramos dimensiones de las etiquetas de entrenamiento.
print("Shape de y_train:", y_train.shape)

# Mostramos dimensiones de las etiquetas de prueba.
print("Shape de y_test:", y_test.shape)
```

### Interpretación del bloque 4

Aquí el dataset queda dividido en dos partes:
- una para **aprender**,
- otra para **evaluar**.

Esto permite medir si el modelo generaliza bien y no solo memoriza los datos.


## Bloque 5. Crear y entrenar el modelo KNN

En este bloque se define el número de vecinos, se crea el clasificador y se entrena con los datos de entrenamiento.


```python
# Definimos el número de vecinos.
k = 5

# Creamos el modelo KNN.
knn = KNeighborsClassifier(
    n_neighbors=k,      # Número de vecinos cercanos
    metric="euclidean"  # Distancia euclidiana
)

# Entrenamos el modelo.
knn.fit(X_train, y_train)

# Confirmamos que el modelo está entrenado.
print(f"Modelo KNN entrenado correctamente con k = {k}")
```

### Interpretación del bloque 5

El modelo ya quedó listo para clasificar nuevas observaciones.

`k = 5` significa que cada nueva flor se clasificará según la mayoría entre sus **5 vecinos más cercanos**.


## Bloque 6. Hacer predicciones y evaluar el modelo

En este bloque se predicen las clases del conjunto de prueba y se calculan métricas de desempeño.


```python
# Predecimos las clases del conjunto de prueba.
y_pred = knn.predict(X_test)

# Calculamos el accuracy.
acc = accuracy_score(y_test, y_pred)

# Calculamos la matriz de confusión.
cm = confusion_matrix(y_test, y_pred)

# Generamos el reporte de clasificación.
report = classification_report(y_test, y_pred, target_names=iris.target_names)

# Mostramos accuracy.
print("Accuracy:", acc)

# Mostramos la matriz de confusión.
print("\nMatriz de confusión:")
print(cm)

# Mostramos el reporte de clasificación.
print("\nReporte de clasificación:")
print(report)
```

### Interpretación del bloque 6

Aquí se observa el desempeño real del modelo.

- **Accuracy** muestra el porcentaje de aciertos.
- **Matriz de confusión** muestra en qué clases se equivoca.
- **Reporte de clasificación** resume precisión, recall y F1-score.

Si la diagonal de la matriz domina, el modelo está clasificando bien.


## Bloque 7. Visualizar fronteras de decisión en 2D

En este bloque se entrena un KNN usando solo dos variables para poder mostrar cómo divide el plano.


```python
# Creamos un nuevo modelo KNN para las dos variables visuales.
knn_vis = KNeighborsClassifier(
    n_neighbors=5,      # Número de vecinos
    metric="euclidean"  # Distancia euclidiana
)

# Entrenamos el modelo con las variables visuales.
knn_vis.fit(X_vis, y)

# Definimos un margen visual.
offset = 0.5

# Calculamos el rango del eje X.
x_min, x_max = X_vis[:, 0].min() - offset, X_vis[:, 0].max() + offset

# Calculamos el rango del eje Y.
y_min, y_max = X_vis[:, 1].min() - offset, X_vis[:, 1].max() + offset

# Construimos una malla de puntos.
x_grid, y_grid = np.meshgrid(
    np.linspace(x_min, x_max, 200),
    np.linspace(y_min, y_max, 200)
)

# Predecimos la clase de cada punto de la malla.
pred_grid = knn_vis.predict(np.c_[x_grid.ravel(), y_grid.ravel()])

# Reorganizamos las predicciones.
pred_grid = pred_grid.reshape(x_grid.shape)

# Creamos la figura.
_, ax = plt.subplots(figsize=(7, 5))

# Pintamos el fondo según la clase predicha.
ax.pcolormesh(
    x_grid,
    y_grid,
    pred_grid,
    cmap="Paired",
    shading="auto"
)

# Superponemos los puntos reales.
ax.scatter(
    X_vis[:, 0],
    X_vis[:, 1],
    c=y,
    cmap="Paired",
    edgecolors="k"
)

# Agregamos título y etiquetas.
ax.set_title("Fronteras de decisión con KNN")
ax.set_xlabel(iris.feature_names[0])
ax.set_ylabel(iris.feature_names[1])

# Mostramos la figura.
plt.tight_layout()
plt.show()
```

### Interpretación del bloque 7

El fondo coloreado representa la clase que el modelo asignaría a una nueva observación en cada región.

Las fronteras muestran cómo KNN divide el espacio según la cercanía entre puntos.  
Por eso suelen verse formas irregulares o muy dependientes de la distribución local.


## Bloque 8. Probar una predicción nueva

En este bloque se crea una nueva flor y se consulta al modelo su clase predicha y las probabilidades por clase.


```python
# Creamos una nueva flor con cuatro características.
nueva_flor = np.array([[5.8, 3.0, 4.35, 1.3]])

# Predecimos la clase de la nueva flor.
pred_nueva = knn.predict(nueva_flor)[0]

# Calculamos probabilidades por clase.
prob_nueva = knn.predict_proba(nueva_flor)[0]

# Mostramos la observación.
print("Nueva flor:", nueva_flor)

# Mostramos la clase predicha.
print("Clase predicha:", pred_nueva)

# Mostramos el nombre de la clase.
print("Nombre de la clase:", iris.target_names[pred_nueva])

# Mostramos probabilidades por clase.
print("Probabilidades por clase:")
for i, prob in enumerate(prob_nueva):
    print(f"  {iris.target_names[i]}: {prob:.4f}")
```

### Interpretación del bloque 8

Este bloque muestra una decisión concreta del modelo.

La clase final depende de los vecinos más cercanos de la nueva flor.  
Las probabilidades permiten ver si la decisión fue clara o si hubo competencia entre varias clases.
